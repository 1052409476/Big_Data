```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# Scatter plot (diagramas de dispersion)
from turtle import color, title


x = np.arange(1, 20)
y = x**2
y1 = x**2 + 5*x

plt.scatter(x,y, s=50, color='red', marker='s', label='x^2')
plt.scatter(x,y1, s=50, color='blue', marker='o', label='x^2 + 5x')
plt.legend(title='Funciones')
plt.title('Tipos de Funciones', fontsize=20)
plt.xlabel('x', color='green')
plt.ylabel('f(x)')
plt.grid()
#plt.xlim(0,5)

plt.text(5, 300, 'este texto', color='brown')

# agegar las coordenas de los marcadores en texto
for val_x in x:
    plt.text(val_x, val_x**2, f'({val_x}, {val_x**2})')
```

```python
# Histograma

from itertools import accumulate


media = 20
std = 3
numeros = np.random.normal(media, std, 10000)

plt.hist(numeros, bins=20, cumulative=False, density=False)
plt.title('Distribucion normal')
plt.xlabel('Numeros')
plt.ylabel('Frecuencia')
```
